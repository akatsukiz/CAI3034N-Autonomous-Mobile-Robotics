**2) Convert BGR to RGB** \
OpenCV reads in images in BGR format (instead of RGB) because when OpenCV was first being developed, BGR color format was popular among camera manufacturers and image software providers

```python
# Method 1
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('C:\Dataset\cute.jpg', 1)
rgb_img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Faster way to convert to RGB
plt.subplot(121),plt.imshow(img),plt.title('BGR');
plt.subplot(122),plt.imshow(rgb_img),plt.title('RGB');
```

```python
# Method 2
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('C:\Dataset\cute.jpg', 1) #1 (color image), 0 (grayscale)
b,g,r = cv2.split(img)           # get b, g, r
rgb_img = cv2.merge([r,g,b])     # rearrange it to r, g, b
plt.subplot(121),plt.imshow(img),plt.title('BGR');
plt.subplot(122),plt.imshow(rgb_img),plt.title('RGB');
```
